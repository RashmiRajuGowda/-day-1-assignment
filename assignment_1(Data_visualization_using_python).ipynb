Question 2: plot a line plot  showing the sales trend in company 1 and 2

```python
days = [1,2,3,4,5,6,7] #days of d week
sales_1 = [160,150,140,145,175,165,180,] #sales of company1
sales_2 = [70,90,160,150,140,145,175] #sales of company2
plt.xlabel("days of week",color = "blue")
plt.ylabel("sales according to week",color = "red")
plt.title("line chart",color="black")
plt.plot(days,sales_1,"bo--",label="line_1",markersize=10)
plt.plot(days,sales_2,"g*--",label="line_2",markersize=10)
plt.legend()
```

Question 3: creat a 3 by 3 subplots: multiple plots

```python
x = [1,2,3,4]
y1 = [4,3,2,1]
y2 = [10,20,30,40]
y3 = [40,30,20,10]
y4 = [1,2,1,2]
y5 = [40,70,90,70]


plt.subplot(3,3,1)
plt.plot(x,y1,"r--")

plt.subplot(3,3,2)
plt.plot(x,y2,"b*-")

plt.subplot(3,3,3)
plt.plot(x,y3,"go--")

plt.subplot(3,3,4)
plt.plot(x,y4,"y--")

plt.subplot(3,3,5)
plt.plot(x,y5,"r*-")

plt.subplot(3,3,6)
plt.plot(x,"bo-")

plt.subplot(3,3,7)
plt.plot(y1,y2,"ro-")

plt.subplot(3,3,8)
plt.plot(y3,y4,"g*--")

plt.subplot(3,3,9)
plt.plot(y1,y4,"b--")
```
